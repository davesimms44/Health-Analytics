```python
#EX 2 A

import matplotlib.pyplot as plt
import seaborn as s
mpg = s.load_dataset('mpg')

mpg.head()

cylcount = mpg['cylinders'].value_counts()
cylcount

#define list of cylinders
cylctdf = pd.DataFrame(cylcount)
cylctdf
cntlist = list(cylctdf.index)
cntlist

cylinders = cntlist

plt.figure(figsize=(12,10))
for cylinder in cylinders: 
    subset = mpg[mpg['cylinders']==cylinder] 
    # this creates 5 dataframes subsetted by 'cylinder', stored within the subset

  # Draw a series of density plots
    s.distplot(subset['mpg'],hist = False, label = cylinder) # remember this is still indented so still part of the loop 
  # this loop goes through 3 times, each time plotting the distplot() density function on top of the figure


plt.title('MPG per Cylinder', fontsize=20)
plt.xlabel('Miles Per Gallon (MPG)', fontsize=20)
plt.ylabel('Density', fontsize=16)
plt.legend()
```

```python
#EX2B
s.displot(data=mpg, x='mpg', row='cylinders', kde=True)
plt.xlabel('Miles Per Gallon(MPG)')
plt.legend()
```

```python
# Ex 12.3

import matplotlib.pyplot as plt
import seaborn as s
import numpy as np
mpg= s.load_dataset('mpg')

mpg.head()


plt.figure(figsize=(5,5))
s.boxplot(data=mpg, x='origin', y='horsepower', orient='v') # orient=h is default, could also be v (vertical)
plt.title("Horsepower by Country of Origin")
plt.xlabel('Country of Origin')
plt.ylabel('Horsepower')
plt.legend()
```

```python
#ex12.4
mpgpie = s.load_dataset('mpg')
mpgpie['cylinders'].value_counts()

cyl_count = list(mpgpie['cylinders'].value_counts())
cyl_count
```

```python
pieLabels = list(mpgpie['cylinders'].value_counts().index)
pieLabels
```

```python
#define Seaborn color palette to use
colors = s.color_palette('pastel')[0:3]
plt.figure(figsize=(12,10))
plt.pie(cyl_count, labels = pieLabels, colors = colors, autopct='%.0f%%', explode=(0,0.3,0,0,0), textprops={'fontsize': 10}) # split one of the pies out
plt.show()
```

```python
#ex12.4

plt.figure(figsize=(12,10))
countryCnt = list(mpgpie['origin'].value_counts())
countryCnt
cntLabels = list(mpgpie['origin'].value_counts().index)
colors = s.color_palette('pastel')[0:3]

plt.pie(countryCnt, labels = cntLabels, colors = colors, autopct='%.0f%%', explode=(0.1,0,0)) # split one of the pies out
plt.title('Origin by Country', fontsize=20)
plt.rcParams['font.size'] = 9.0
plt.legend()
plt.show()
```
