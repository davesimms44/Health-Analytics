```python
#1) From assignment 1, import the provided xlsx files (the same ones you imported into MSAccess) into Jupyter: 
#Comment2, FallTable2, InterviewTable2, MedicalHistory2, SurgicalHistory2

Comment2 = pd.read_excel("Comment2.xlsx")
Interview2 = pd.read_excel("InterviewTable2.xlsx")
Fall2 = pd.read_excel("FallTable2.xlsx")
MedHx2 = pd.read_excel("MedicalHistory2.xlsx")
SurgHx2 = pd.read_excel("SurgicalHistory2.xlsx")
```

```python
# 2) Sort each of them by SubjectID
# Could sort by multiple variables, and define if ascending or descending using list 
# matching the same order of variables
# na_position forces NA values to first or last 

Comment2.sort_values(by=['SubjectID'], na_position="last")
Interview2.sort_values(by=['SubjectID'], na_position="last")
Fall2.sort_values(by=['SubjectID'], na_position="last")
MedHx2.sort_values(by=['SubjectID'], na_position="last")
SurgHx2.sort_values(by=['SubjectID'], na_position="last")
```

```python
#3) Merge all of them together in an inner join as a new dataframe
TablesJoin = Comment2.merge(Interview2, on="SubjectID", how="inner")
TablesJoin = TablesJoin.merge(Fall2, on="SubjectID", how="inner")
TablesJoin = TablesJoin.merge(MedHx2, on="SubjectID", how="inner")
TablesJoin = TablesJoin.merge(SurgHx2, on="SubjectID", how="inner")
print(TablesJoin)
```
